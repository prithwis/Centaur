Minimal LLM call (proof of life)

This is the absolute minimum Centaur neural spine.

```python
from openai import OpenAI

client = OpenAI()

def llm_call(system_prompt, user_prompt, model="gpt-4o-mini"):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
        temperature=0.2,
    )
    return response.choices[0].message.content.strip()
```

The Centaur adjudicator (first incarnation)

```python
SYSTEM_PROMPT = """
You are a strategic adjudicator in a matrix strategy game.
Your task is to rule on player actions conservatively.

Principles:
- Do not invent rules.
- Do not assume success.
- Evaluate plausibility and proportional impact.
- Justify your ruling briefly and clearly.
"""

def centaur_adjudicate(
    human_move: str,
    game_state: dict,
    human_override: str = None
):
    context = f"""
Game State:
{game_state}

Player Move:
{human_move}

Respond with:
1. Outcome
2. Reasoning
"""

    llm_ruling = llm_call(SYSTEM_PROMPT, context)

    final_ruling = human_override if human_override else llm_ruling

    return {
        "human_move": human_move,
        "llm_ruling": llm_ruling,
        "final_ruling": final_ruling
    }
```

First Run - Is the Centaur awake?

```python
game_state = {
    "turn": 1,
    "resources": {
        "Power": 100,
        "Legitimacy": 100
    }
}

result = centaur_adjudicate(
    human_move="Conduct covert cyber operations to undermine enemy command morale",
    game_state=game_state
)

print("=== LLM RULING ===")
print(result["llm_ruling"])
```
